```python
# Data Leakage Check

# Columns that may cause data leakage
print("Checking potential leakage columns:")
print(df[['pickup_datetime', 'dropoff_datetime', 'trip_duration']].head())

# Drop the column that leaks the target information
df = df.drop(columns=['dropoff_datetime'])

print("\nColumn 'dropoff_datetime' removed to prevent data leakage.")

# Verify remaining columns
print("\nRemaining Columns:")
print(df.columns)
```

```python
# Feature Engineering

import numpy as np

# Converting pickup_datetime to datetime format
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])


# 1. Pickup hour
df['pickup_hour'] = df['pickup_datetime'].dt.hour


# 2. Pickup day of week
df['pickup_day_of_week'] = df['pickup_datetime'].dt.dayofweek


# 3. Pickup month
df['pickup_month'] = df['pickup_datetime'].dt.month


# 4. Trip distance (Haversine formula)

def haversine_distance(lat1, lon1, lat2, lon2):
    
    R = 6371  
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    return R * c


df['trip_distance'] = haversine_distance(
    df['pickup_latitude'],
    df['pickup_longitude'],
    df['dropoff_latitude'],
    df['dropoff_longitude']
)


# Display new features
print("Below are the New Features Created:")
print(df[['pickup_hour','pickup_day_of_week','pickup_month','trip_distance']].head())
```

```python
# Data Cleaning

# 1. Removing extreme trip duration values
df = df[(df['trip_duration'] > 60) & (df['trip_duration'] < 7200)]


# 2. Removing invalid passenger counts
df = df[(df['passenger_count'] > 0) & (df['passenger_count'] <= 6)]


# 3. Removing invalid pickup coordinates
df = df[(df['pickup_latitude'] > 40.5) & (df['pickup_latitude'] < 41.0)]
df = df[(df['pickup_longitude'] > -74.5) & (df['pickup_longitude'] < -73.5)]


# 4. Removing invalid dropoff coordinates
df = df[(df['dropoff_latitude'] > 40.5) & (df['dropoff_latitude'] < 41.0)]
df = df[(df['dropoff_longitude'] > -74.5) & (df['dropoff_longitude'] < -73.5)]


# Checking dataset shape after cleaning
print("Dataset shape after cleaning:", df.shape)
```

```python
# Outlier Detection

plt.figure(figsize=(8,5))
sns.boxplot(x=df['trip_duration'])
plt.title("Trip Duration After Cleaning (Outlier Check)")
plt.xlabel("Trip Duration (seconds)")
plt.show()
```

```python
# Encoding Categorical Variables and Feature Selection

# 1. Encoding 'store_and_fwd_flag'
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].map({'N': 0, 'Y': 1})

# 2. Dropping unnecessary columns
df = df.drop(columns=['id', 'pickup_datetime'])

# 3. Final check
print("Final Columns for Modeling:")
print(df.columns)

print("\nDataset shape after preprocessing:")
print(df.shape)

print("\nFirst 5 rows of final dataset:")
display(df.head())
```
